```python
from pyspark import SparkContext
import time
NodeList = {}
class Node:
    def __init__(self,id):  
        self.id = id
        self.Children = []
        self.minId = id

        
    def AddChild(self,cid):
        self.Children.append(cid)
        
    @staticmethod
    def get_node(id):
        global NodeList
        if id not in NodeList:
            NodeList[id] = Node(id)
        return NodeList[id]

    

class Processor:
    def __init__(self,fname):
        self.Nodes = {}
        self.edges = []
        ifile = open(fname)
        c = 0
        for l in ifile:
            c += 1
            if c > 100000000:
                break
            l = l.strip()
            if len(l) < 1:
                continue
            l = l.split(' ')
            fid,sid = int(l[0]),int(l[1])
            fnode = Node.get_node(fid)
            snode = Node.get_node(sid)
            fnode.AddChild(sid)
            snode.AddChild(fid)
            
    def update_vertices(self,r):
        global NodeList
        vertex_updated = False
        for x in r:
            if NodeList[x[0]].minId != x[1]:
                vertex_updated = True
                

            NodeList[x[0]].minId = x[1]
        return vertex_updated
        
    def start(self):
        global NodeList
        nids = list(NodeList.keys())
        sc = SparkContext.getOrCreate()
        nids = sc.parallelize(nids)
        while True:
            start_time = time.time()
            mapper = Mapper()
            m = mapper.map(nids,NodeList)
            reducer = Reducer()
            r = reducer.reduce(m)
            if not p.update_vertices(r):
                minids = list(set([NodeList[nid].minId for nid in NodeList]))
                print(len(minids))
                return
            end_time = time.time()
            print(len(list(set([NodeList[nid].minId for nid in NodeList]))),end_time - start_time)





class Mapper:
    def __init__(self):
        pass
    
    def map_func(self,nid,Nodes):
        return [Nodes[cid].minId for cid in Nodes[nid].Children] + [Nodes[nid].minId]
        
    
    def map(self,nids,Nodes):
        return nids.map(lambda nid:(nid,self.map_func(nid,Nodes)))
        
    


class Reducer:
    def __init__(self):
        pass
    
    def reduce(self,m):
        m = m.map(lambda x:(x[0],min(x[1])))
        return m.collect()

p = Processor('undirect.graph')
p.start()
print("Done")
```
