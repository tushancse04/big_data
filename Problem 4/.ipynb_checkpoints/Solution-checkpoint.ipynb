```python
import re
import pickle
from pyspark import SparkContext

products = {}
queries = []
class Product:
    def __init__(self,id):
        self.id = id
        self.total_reviews = 0
        self.min_rate = 0
        self.categories = []
        
    @staticmethod
    def get_prod(id):
        global products
        if id not in products:
            products[id] = Product(id)
        return products[id]
    
    def AddCategory(self,catId):
        self.categories.append(catId)

class FileProcessor:
    def __init__(self,fname,qfile):
        lstr = ""
        lines = open(fname,encoding="utf8").readlines()#[0:1000]
        print('read done')

        count = 0
        prod = ""
        for l in lines:
            if "Id:" not in l:
                prod += l + " "
                continue


            if 'ASIN:' not in prod:
                prod = l
                continue

            id = int(prod.split('ASIN:')[0].split('Id:')[1])
            product = Product.get_prod(id)
            if 'categories:' not in prod:
                continue
            prod = prod.split('categories:')[1]
            if 'reviews:' not in prod:
                prod = l
                continue
                
            prod = prod.split('reviews:')
            categories = prod[0].split('[')
            for cat in categories:
                if ']' in cat:
                    cat = cat.split(']')[0]
                    product.AddCategory(cat)
                    
            
            prod = prod[1].split('  downloaded:')
            total_reviews = int(prod[0].split('total: ')[1])
            product.total_reviews = total_reviews
            ratings = prod[1].split('avg rating:')[1]
            ratings = ratings.split('rating:')[1:]
            min_rate = 5
            for r in ratings:
                r = int(r.split('  votes')[0])
                if r < min_rate:
                    min_rate = r
            product.min_rate = min_rate
            prod = l
            count += 1

                
class Mapper:
    def __init__(self):
        pass
    
    def map_func(self,x,products):

        i = x[0][0]
        x = x[0][1]
        x = x.strip().split(' ')
        cid = x[0]
        n = int(x[1])
        tot_reviews = int(x[2])
        min_rate = int(x[3])
 
        return (i,[[products[id].min_rate,products[id].total_reviews,id] \
                for id in products if cid in products[id].categories \
                and products[id].min_rate >= min_rate and products[id].total_reviews >= tot_reviews])
        
    
    def map(self,query,products):
        return query.map(lambda x:self.map_func(x,products))
                 
class Reducer:
    def __init__(self):
        pass
    
    def red_func(self,x,query):
        n = int(query[x[0]][0][1].split(' ')[1])
        l = sorted(x[1], key = lambda x: (-x[0],-x[1], x[2]))[0:n]
        l = [x[2] for x in l]
        return (x[0],l)
    
    def reduce(self,m,query):
        return m.map(lambda x:self.red_func(x,query))
    
p = FileProcessor('amazon-meta.txt','query')
pickle.dump(products,open("products.p","wb"))
#products = pickle.load(open( 'products.p', "rb" ) )
sc = SparkContext.getOrCreate()
query = sc.textFile("query")
query = query.collect()
for i in range(len(query)):
    query[i] = [(i,query[i])]

query = sc.parallelize(query)
mapper = Mapper()
m = mapper.map(query,products)
reducer = Reducer()
r = reducer.reduce(m,query.collect())
print(r.collect())
        
```
